# Q.5 What is the highest Monthly earning channel type

```python
sns.boxplot(x = "channel_type", y = "highest_monthly_earnings", data = df)
plt.title("Boxplot of Highest Monthly Earnings by Channel Type")
plt.xlabel("Channel Type")
plt.ylabel("Highest Monthly Earnings")
plt.show()
```

# Q.6 what is the total of subscribers by category and country

```python
df.groupby(["category","Country"])["subscribers"].sum().unstack().plot(kind = "bar",stacked = True)
plt.title("Total Subscribers by Category and Country")
plt.xlabel("Category")
plt.ylabel("Total Subscribers")
plt.legend(title = "Country",bbox_to_anchor = (1.1,1.0),loc = "upper left")
plt.show()
```

# Q.7 what is the gross territary education enrollment  % vs subscribers

```python
plt.scatter(df["Gross tertiary education enrollment (%)"],df["subscribers"],alpha = 0.5)
plt.title("Gross tertiary education enrollment (%) vs subscribers")
plt.xlabel("Gross tertiary education enrollment (%)")
plt.ylabel("subscribers")
plt.show()
```

# Q.8 What is the average rate by channel type

```python
df.groupby("channel_type")["Unemployment rate"].mean().head().sort_values(ascending =False).plot(kind = "bar")
plt.title("Average unemployment rate by Channel Type")
plt.xlabel("Channel Type")
plt.ylabel("Unemployment rate")
plt.show()
```

# Q.9 what is the average ratio of subscribers by created year and channel type

```python
df.groupby(["created_year","channel_type"]) ["subscribers"].mean().unstack().tail().plot(kind = "bar", width = 1)
plt.title("Average subscribers by created year and channel type")
plt.xlabel("created year")
plt.ylabel("subscribers")
plt.xticks(rotation = 45)
plt.legend(title = "channel type",bbox_to_anchor = (1.1,1.0),loc = "upper left")
plt.show()
```

# Q.10 What is the subscribers growth overtime

```python
df.groupby(["created_year","created_month"])["subscribers"].sum().unstack().head().plot(kind = "area",stacked = True)
plt.title("Subscribers growth overtime")
plt.xlabel("Year")
plt.ylabel("Total subscribers")
plt.legend(title = "created month",bbox_to_anchor = (1.1,1.0),loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```

# Q.11 what is the trend of highest yearly earnings by created year and country

```python
df.groupby(["created_year","Country"])["highest_yearly_earnings"].mean().unstack().head(10).plot(kind = "line",marker = "o")
plt.title("Trend of highest yearly earnings by created year and country")
plt.xlabel("created year")
plt.ylabel("highest yearly earnings")
plt.legend(title = "Country",bbox_to_anchor = (1.1,1.0),loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```
